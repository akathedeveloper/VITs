```python
import os
import math
import time
import random
from tqdm import tqdm
import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
import torchvision.transforms as T
from torch.utils.data import DataLoader


SEED = 42
torch.manual_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed_all(SEED)

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Device:", device)


# Utilities

def count_params(model):
    total = sum(p.numel() for p in model.parameters())
    trainable = sum(p.numel() for p in model.parameters() if p.requires_grad)
    return total, trainable

def try_flops(model, input_size=(1,3,32,32)):
    try:
        from thop import profile
        model_cpu = model.to('cpu')
        dummy = torch.randn(*input_size)
        macs, params = profile(model_cpu, inputs=(dummy,), verbose=False)
        flops = 2 * macs
        model.to(device)
        return flops
    except Exception:
        return None

# Exponential Moving Average (EMA) for model weights
class ModelEMA:
    def __init__(self, model, decay=0.9999):
        self.decay = decay
        self.shadow = {}
        self.collected = False
        for name, p in model.named_parameters():
            if p.requires_grad:
                self.shadow[name] = p.detach().cpu().clone()

    def update(self, model):
        for name, p in model.named_parameters():
            if p.requires_grad:
                new = p.detach().cpu().clone()
                self.shadow[name].mul_(self.decay).add_(new, alpha=1.0 - self.decay)

    def apply_shadow(self, model):
        for name, p in model.named_parameters():
            if p.requires_grad:
                p.data.copy_(self.shadow[name].to(p.device))


# Building blocks
# DropPath implementation (stochastic depth)
class DropPath(nn.Module):
    def __init__(self, drop_prob=0.0):
        super().__init__()
        self.drop_prob = drop_prob
    def forward(self, x):
        if not self.training or self.drop_prob == 0.:
            return x
        keep_prob = 1 - self.drop_prob
        # shape: (batch, 1, 1)
        shape = (x.shape[0],) + (1,) * (x.ndim - 1)
        rand = torch.empty(shape, dtype=x.dtype, device=x.device).bernoulli_(keep_prob)
        return x.div(keep_prob) * rand

# Depthwise conv 
class DWConv(nn.Module):
    def __init__(self, dim, kernel=3):
        super().__init__()
        self.dw = nn.Conv2d(dim, dim, kernel_size=kernel, padding=kernel//2, groups=dim, bias=False)
        self.bn = nn.BatchNorm2d(dim)
        self.act = nn.ReLU(inplace=True)
    def forward(self, x, H, W):
        # x: [B, N, C] -> convert to [B, C, H, W]
        B, N, C = x.shape
        x2 = x.transpose(1,2).reshape(B, C, H, W)
        x2 = self.dw(x2)
        x2 = self.bn(x2)
        x2 = self.act(x2)
        x2 = x2.reshape(B, C, -1).transpose(1,2)
        return x2

# FFN (MLP) with 2x expansion 
class FFN(nn.Module):
    def __init__(self, dim, mlp_ratio=2.0, dropout=0.0):
        super().__init__()
        hidden = int(dim * mlp_ratio)
        self.fc1 = nn.Linear(dim, hidden)
        self.act = nn.GELU()
        self.fc2 = nn.Linear(hidden, dim)
        self.drop = nn.Dropout(dropout)
    def forward(self, x):
        x = self.fc1(x)
        x = self.act(x)
        x = self.drop(x)
        x = self.fc2(x)
        x = self.drop(x)
        return x

# Cascaded Group Attention (CGA) 
class CascadedGroupAttention(nn.Module):
    def __init__(self, dim, num_heads=4, qk_ratio=0.25):
        super().__init__()
        assert dim % num_heads == 0, "dim must be divisible by num_heads"
        self.dim = dim
        self.h = num_heads
        self.head_dim = dim // num_heads
        self.qk_dim = max(1, int(self.head_dim * qk_ratio))

        # per-head small Q/K projections and full V projections (per-head)
        self.q_projs = nn.ModuleList([nn.Linear(self.head_dim, self.qk_dim, bias=False) for _ in range(self.h)])
        self.k_projs = nn.ModuleList([nn.Linear(self.head_dim, self.qk_dim, bias=False) for _ in range(self.h)])
        self.v_projs = nn.ModuleList([nn.Linear(self.head_dim, self.head_dim, bias=False) for _ in range(self.h)])

        self.prev_projs = nn.ModuleList([nn.Linear(self.head_dim, self.head_dim, bias=False) for _ in range(self.h)])

        self.out_proj = nn.Linear(dim, dim)

    def forward(self, x, H, W):
        # x: [B, N, C]
        B, N, C = x.shape
        x_split = x.view(B, N, self.h, self.head_dim).permute(2,0,1,3)

        head_outputs = []
        prev_out = None
        for j in range(self.h):
            xj = x_split[j]  # [B, N, head_dim]
            if prev_out is not None:
                add = self.prev_projs[j](prev_out)
                xj = xj + add

            q = self.q_projs[j](xj)   # [B, N, qk_dim]
            k = self.k_projs[j](xj)   # [B, N, qk_dim]
            v = self.v_projs[j](xj)   # [B, N, head_dim]

            attn = torch.matmul(q, k.transpose(-2, -1)) / math.sqrt(max(1.0, float(self.qk_dim)))
            attn = attn.softmax(dim=-1)
            out_j = torch.matmul(attn, v)  # [B, N, head_dim]

            head_outputs.append(out_j)
            prev_out = out_j.detach()

        out = torch.cat(head_outputs, dim=-1)
        return self.out_proj(out)

# Sandwich block: FFN -> DWConv -> CGA -> DWConv -> FFN
class SandwichBlock(nn.Module):
    def __init__(self, dim, num_heads=4, mlp_ratio=2.0, drop_path=0.0, qk_ratio=0.25):
        super().__init__()
        self.norm1 = nn.LayerNorm(dim)
        self.ffn1 = FFN(dim, mlp_ratio)
        self.norm2 = nn.LayerNorm(dim)
        self.dw1 = DWConv(dim)
        self.norm3 = nn.LayerNorm(dim)
        self.attn = CascadedGroupAttention(dim, num_heads=num_heads, qk_ratio=qk_ratio)
        self.norm4 = nn.LayerNorm(dim)
        self.dw2 = DWConv(dim)
        self.norm5 = nn.LayerNorm(dim)
        self.ffn2 = FFN(dim, mlp_ratio)
        self.drop_path = DropPath(drop_path) if drop_path > 0. else nn.Identity()

    def forward(self, x, H, W):
        # x: [B, N, C]
        y = self.norm1(x)
        y = self.ffn1(y)
        x = x + self.drop_path(y)

        y = self.norm2(x)
        y = self.dw1(y, H, W)
        x = x + self.drop_path(y)

        y = self.norm3(x)
        y = self.attn(y, H, W)
        x = x + self.drop_path(y)

        y = self.norm4(x)
        y = self.dw2(y, H, W)
        x = x + self.drop_path(y)

        y = self.norm5(x)
        y = self.ffn2(y)
        x = x + self.drop_path(y)
        return x

# Tiny EfficientViT
class TinyEfficientViT(nn.Module):
    def __init__(self,
                 num_classes=10,
                 stage_channels=(64, 128, 192),
                 stage_depths=(1, 2, 3),
                 stage_heads=(2, 4, 4),
                 mlp_ratio=2.0,
                 qk_ratio=0.25,
                 drop_path_rate=0.1):
       
        super().__init__()
        assert len(stage_channels) == len(stage_depths) == len(stage_heads)

        C0 = stage_channels[0]
        self.stem = nn.Sequential(
            nn.Conv2d(3, C0, kernel_size=3, stride=1, padding=1, bias=False),
            nn.BatchNorm2d(C0),
            nn.ReLU(inplace=True),
            nn.Conv2d(C0, C0, kernel_size=3, stride=2, padding=1, bias=False),
            nn.BatchNorm2d(C0),
            nn.ReLU(inplace=True),
        )

        self.stage_channels = stage_channels
        self.stage_blocks = nn.ModuleList()
        self.stage_proj = nn.ModuleList()

        total_blocks = sum(stage_depths)
        dp_rates = list(np.linspace(0.0, drop_path_rate, total_blocks))
        dp_iter = 0

        for idx, (C, depth, heads) in enumerate(zip(stage_channels, stage_depths, stage_heads)):
            proj = nn.Sequential(nn.Conv2d(C, C, kernel_size=1, stride=1, bias=False),
                                 nn.BatchNorm2d(C),
                                 nn.ReLU(inplace=True))
            self.stage_proj.append(proj)

            blocks = nn.ModuleList()
            for b in range(depth):
                blocks.append(SandwichBlock(dim=C, num_heads=heads, mlp_ratio=mlp_ratio,
                                           drop_path=dp_rates[dp_iter], qk_ratio=qk_ratio))
                dp_iter += 1
            self.stage_blocks.append(blocks)

            if idx < len(stage_channels) - 1:
                nextC = stage_channels[idx + 1]
                down = nn.Sequential(
                    nn.Conv2d(C, nextC, kernel_size=3, stride=2, padding=1, bias=False),
                    nn.BatchNorm2d(nextC),
                    nn.ReLU(inplace=True),
                )
                setattr(self, f"down_{idx}", down)

        self.global_pool = nn.AdaptiveAvgPool2d(1)
        self.head = nn.Linear(stage_channels[-1], num_classes)

        self._init_weights()

    def _init_weights(self):
        for m in self.modules():
            if isinstance(m, nn.Conv2d):
                nn.init.kaiming_normal_(m.weight, mode='fan_out', nonlinearity='relu')
            elif isinstance(m, nn.Linear):
                nn.init.trunc_normal_(m.weight, std=0.02)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)
            elif isinstance(m, nn.BatchNorm2d):
                nn.init.ones_(m.weight)
                nn.init.zeros_(m.bias)

    def forward(self, x):
        B = x.shape[0]
        x = self.stem(x)  # [B, C1, H/2, W/2]
        for i, blocks in enumerate(self.stage_blocks):
            B, C, H, W = x.shape
            for blk in blocks:
                tokens = x.reshape(B, C, -1).permute(0, 2, 1)  # [B, N, C]
                tokens = blk(tokens, H, W)
                x = tokens.permute(0, 2, 1).reshape(B, C, H, W)
            if i < len(self.stage_blocks) - 1:
                x = getattr(self, f"down_{i}")(x)
                x = self.stage_proj[i+1](x)

        x = self.global_pool(x).reshape(B, -1)
        x = self.head(x)
        return x

# CIFAR-10

def get_cifar10_loaders(batch_size=128, num_workers=4, use_aug=True):
    mean = (0.4914, 0.4822, 0.4465)
    std = (0.2471, 0.2435, 0.2616)

    if use_aug:
        train_transforms = T.Compose([
            T.RandAugment(num_ops=2, magnitude=9),   
            T.RandomCrop(32, padding=4),
            T.RandomHorizontalFlip(),
            T.ToTensor(),
            T.Normalize(mean, std),
        ])
    else:
        train_transforms = T.Compose([
            T.RandomCrop(32, padding=4),
            T.RandomHorizontalFlip(),
            T.ToTensor(),
            T.Normalize(mean, std),
        ])

    test_transforms = T.Compose([
        T.ToTensor(),
        T.Normalize(mean, std),
    ])

    train_set = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=train_transforms)
    test_set = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=test_transforms)

    train_loader = DataLoader(train_set, batch_size=batch_size, shuffle=True, num_workers=num_workers, pin_memory=True)
    test_loader = DataLoader(test_set, batch_size=batch_size, shuffle=False, num_workers=num_workers, pin_memory=True)

    return train_loader, test_loader


# Train functions
def train_one_epoch(model, loader, optimizer, device, loss_fn):
    model.train()
    running_loss = 0.0
    top1_correct = 0
    top5_correct = 0
    total = 0
    pbar = tqdm(loader, leave=False)
    for xb, yb in pbar:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        out = model(xb)
        loss = loss_fn(out, yb)
        loss.backward()
        optimizer.step()

        bs = xb.size(0)
        running_loss += loss.item() * bs
        total += bs
        preds = out.topk(5, dim=1)[1]
        top1_correct += (preds[:,0] == yb).sum().item()
        top5_correct += sum([1 if yb[i].item() in preds[i].tolist() else 0 for i in range(bs)])

    avg_loss = running_loss / total
    top1 = top1_correct / total
    top5 = top5_correct / total
    return avg_loss, top1, top5

def evaluate(model, loader, device, loss_fn):
    model.eval()
    running_loss = 0.0
    top1_correct = 0
    top5_correct = 0
    total = 0
    with torch.no_grad():
        for xb, yb in loader:
            xb, yb = xb.to(device), yb.to(device)
            out = model(xb)
            loss = loss_fn(out, yb)
            bs = xb.size(0)
            running_loss += loss.item() * bs
            total += bs
            preds = out.topk(5, dim=1)[1]
            top1_correct += (preds[:,0] == yb).sum().item()
            top5_correct += sum([1 if yb[i].item() in preds[i].tolist() else 0 for i in range(bs)])
    avg_loss = running_loss / total
    top1 = top1_correct / total
    top5 = top5_correct / total
    return avg_loss, top1, top5


# Plotting
def plot_history(history, out_dir='outputs'):
    os.makedirs(out_dir, exist_ok=True)
    epochs = len(history['train_loss'])
    plt.figure(figsize=(6,4))
    plt.plot(range(1, epochs+1), history['train_loss'], label='train_loss')
    plt.plot(range(1, epochs+1), history['val_loss'], label='val_loss')
    plt.title('Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)
    plt.tight_layout(); plt.savefig(os.path.join(out_dir, 'loss.png')); plt.show()

    plt.figure(figsize=(6,4))
    plt.plot(range(1, epochs+1), np.array(history['train_top1'])*100, label='train_top1')
    plt.plot(range(1, epochs+1), np.array(history['val_top1'])*100, label='val_top1')
    plt.plot(range(1, epochs+1), np.array(history['val_top5'])*100, label='val_top5')
    plt.title('Accuracy (%)'); plt.xlabel('Epoch'); plt.ylabel('Accuracy (%)'); plt.legend(); plt.grid(True)
    plt.tight_layout(); plt.savefig(os.path.join(out_dir, 'accuracy.png')); plt.show()


# Main
def run_experiment(
    batch_size=128,
    epochs=150,
    lr=3e-4,
    weight_decay=0.05,
    use_ema=False,
    patience=20,            #
    use_augment=True,
    label_smoothing=0.0,
):

    stage_channels = (64, 120, 192)   
    stage_depths   = (1, 2, 3)
    stage_heads    = (2, 4, 4)
    mlp_ratio = 2.0
    qk_ratio = 0.25
    drop_path_rate = 0.1

    model = TinyEfficientViT(num_classes=10,
                             stage_channels=stage_channels,
                             stage_depths=stage_depths,
                             stage_heads=stage_heads,
                             mlp_ratio=mlp_ratio,
                             qk_ratio=qk_ratio,
                             drop_path_rate=drop_path_rate).to(device)

    total_params, trainable = count_params(model)
    print(f"Model params: {total_params:,} ({total_params/1e6:.3f}M) | Trainable: {trainable:,}")
    flops = try_flops(model)
    if flops is not None:
        print(f"Approx FLOPs (per-image): {flops:.2e}")

    train_loader, test_loader = get_cifar10_loaders(batch_size, use_aug=use_augment)

    optimizer = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
    loss_fn = nn.CrossEntropyLoss(label_smoothing=label_smoothing).to(device)

    ema = ModelEMA(model) if use_ema else None

    history = {'train_loss': [], 'train_top1': [], 'train_top5': [], 'val_loss': [], 'val_top1': [], 'val_top5': []}
    best_val = 0.0
    best_val_loss = 1e9
    best_epoch = -1
    os.makedirs('outputs', exist_ok=True)
    wait = 0

    print("Starting training ...")
    for ep in range(1, epochs+1):
        t0 = time.time()
        train_loss, train_top1, train_top5 = train_one_epoch(model, train_loader, optimizer, device, loss_fn)
        val_loss, val_top1, val_top5 = evaluate(model, test_loader, device, loss_fn)
        if ema is not None:
            ema.update(model)
        scheduler.step()

        history['train_loss'].append(train_loss); history['train_top1'].append(train_top1); history['train_top5'].append(train_top5)
        history['val_loss'].append(val_loss); history['val_top1'].append(val_top1); history['val_top5'].append(val_top5)

        # early stopping
        improved = False
        if val_top1 > best_val or (val_top1 == best_val and val_loss < best_val_loss):
            improved = True
            best_val = val_top1
            best_val_loss = val_loss
            best_epoch = ep
            wait = 0
            # save state
            save_path = os.path.join('outputs', 'best_efficientvit_cifar10.pth')
            torch.save({
                'epoch': ep,
                'model_state': model.state_dict(),
                'optimizer_state': optimizer.state_dict(),
                'val_top1': val_top1,
                'val_loss': val_loss
            }, save_path)
        else:
            wait += 1

        print(f"E{ep:03d}/{epochs} | TL {train_loss:.4f} | VL {val_loss:.4f} | T1 {train_top1*100:5.2f}% | V1 {val_top1*100:5.2f}% | V5 {val_top5*100:5.2f}% | wait {wait}")
        # early stopping check
        if wait >= patience:
            print(f"⏹ Early stopping triggered at epoch {ep} (no top-1 improvement for {patience} epochs).")
            break

    # final load best weights (already saved)
    final_state = torch.load(os.path.join('outputs', 'best_efficientvit_cifar10.pth'), map_location=device)
    model.load_state_dict(final_state['model_state'])

    print("\n--- Final Validation Metrics (best saved) ---")
    val_loss, val_top1, val_top5 = evaluate(model, test_loader, device, loss_fn)
    print(f"Final Val Loss: {val_loss:.4f}")
    print(f"Final Val Top-1 Acc: {val_top1*100:.2f}%")
    print(f"Final Val Top-5 Acc: {val_top5*100:.2f}%")
    print(f"Best epoch: {best_epoch}")

    plot_history(history, out_dir='outputs')
    return model, history


if __name__ == '__main__':
    model, history = run_experiment(
        batch_size=128,
        epochs=150,
        lr=3e-4,
        weight_decay=0.05,
        use_ema=False,        
        patience=20,         
        use_augment=True,
        label_smoothing=0.0,  
    )
```
